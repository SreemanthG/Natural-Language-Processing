```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem import SnowballStemmer
```

```python
porter_stemmer = PorterStemmer()
lanca_stemmer = LancasterStemmer()
sb_stemmer = SnowballStemmer("english",)

word_data = "The Fall of Kampala was a battle during the Uganda–Tanzania War in April 1979, in which the combined forces of Tanzania\
and the Uganda National Liberation Front (UNLF) attacked and captured the Ugandan capital, Kampala.\
Tanzanian forces were repulsing an invasion launched by Ugandan President Idi Amin (pictured).\
After routing the Ugandans and their Libyan allies in Entebbe, the Tanzanians moved on Kampala." 
sent_tokens = nltk.sent_tokenize(word_data)
for sent in sent_tokens:
    nltk_tokens = nltk.word_tokenize(word_data)
    print(nltk_tokens)
#Next find the roots of the word
print ('\n**Porter**\n')
for w_port in nltk_tokens:
      print (w_port,porter_stemmer.stem(w_port))
print ('\n**Lanca**\n')
for w_lanca in nltk_tokens:
      print (w_lanca,lanca_stemmer.stem(w_lanca))
print ('\n**Snowball**\n')
for w_snow in nltk_tokens:
      print (w_snow,sb_stemmer.stem(w_snow))
```
